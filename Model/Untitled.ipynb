```python
books=books[['ISBN', 'Book-Title', 'Book-Author', 'Year-Of-Publication', 'Publisher','Image-URL-L']]
```

```python
books.rename(columns={
    "Book-Title":"Title",
        "Book-Author":"Author",
            "Year-Of-Publication":"Year",
                "Image-URL-L":"Image-URl",
},inplace=True
)
```

```python
ratings=pd.read_csv('dataset/Ratings.csv' ,encoding='latin-1',on_bad_lines = "skip")
```

```python
ratings.head()
```

```python
ratings.rename(columns={ 
                "Book-Rating":"Rating",
},inplace=True
)
ratings.head()
```

```python
ratings["User-ID"].value_counts()
```

```python
x = ratings["User-ID"].value_counts() > 200
```

```python
x[x].shape
```

```python
y=x[x].index
```

```python
y
```

```python
ratings=ratings[ratings["User-ID"].isin(y)]
```

```python
ratings.head()
```

```python
ratings.shape
```

```python
ratings_with_books = ratings.merge(books, on='ISBN')
```

```python
ratings_with_books.head()
```

```python
ratings_with_books.head()
```

```python
number_rating = ratings_with_books.groupby('Title')['Rating'].count().reset_index()
```

```python
number_rating.head()
```

```python
number_rating.rename(columns={'Rating':'num_of_rating'},inplace=True)
```

```python
number_rating.head()
```

```python
final_rating = ratings_with_books.merge(number_rating, on='Title')
```

```python
final_rating.head()
```

```python
final_rating.shape
```

```python
final_rating = final_rating[final_rating['num_of_rating'] >= 50]
```

```python
final_rating.head()
```

```python
final_rating.shape
```

```python
final_rating.rename(columns={
                "Image-URl":"ImageURl",
},inplace=True
)
```

```python
final_rating.head()
```

```python
final_rating.drop_duplicates(['User-ID','Title'],inplace=True)
```

```python
final_rating.shape
```

```python
book_pivot = final_rating.pivot_table(columns='User-ID', index='Title', values= 'Rating')
book_pivot
```

```python
book_pivot.shape
```

```python
book_pivot.fillna(0, inplace=True)
```

```python
book_pivot
```

```python
from scipy.sparse import csr_matrix
```

```python
book_sparse = csr_matrix(book_pivot)
type(book_sparse)
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm= 'brute')
```

```python
model.fit(book_sparse)
```

```python
distance, suggestion = model.kneighbors(book_pivot.iloc[237,:].values.reshape(1,-1), n_neighbors=6 )
```

```python
def recommend_book(book_name):
    book_id = np.where(book_pivot.index == book_name)[0][0]
    distance, suggestion = model.kneighbors(book_pivot.iloc[book_id,:].values.reshape(1,-1), n_neighbors=6 )
    
    for i in range(len(suggestion)):
            books = book_pivot.index[suggestion[i]]
            for j in books:
                if j == book_name:
                    print(f"You searched '{book_name}'\n")
                    print("The suggestion books are: \n")
                else:
                    print(j)
```

```python
book_name = "Harry Potter and the Chamber of Secrets (Book 2)"
recommend_book(book_name)
```
